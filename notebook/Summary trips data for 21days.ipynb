```python
def find_days(driverid,days):
    count=0
    for j in days:
        if driverid in j:
            count+=1
    
    return count
    
    
```

```python
for index, row in df_driver_summary.iterrows():
    day_count=find_days(row[0],days)
    df_driver_summary.at[index,'number of days']=day_count
    
```
