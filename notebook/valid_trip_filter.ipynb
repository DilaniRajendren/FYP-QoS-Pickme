```python
df=pd.read_csv(path)
```

```python
df_filtered_avg_points = df[df['AverageP'] <= 15]
```

```python
df_filtered_avg_points
```

```python
df_filtered_avg_points = df_filtered_avg_points[df_filtered_avg_points["distance"] > 0.05]
df_filtered_avg_points = df_filtered_avg_points[df_filtered_avg_points["duration"] > 0]
```

```python
df_filtered_avg_points
```

```python
df_filtered_avg_points.to_csv("validtrips_date=2019-11-08.csv")
```
